# Ass2

```python
Input = input("Sample numbers :  ")
List = Input.split(",")
odd = 0
even = 0
for item in List:
    if int(item)%2 == 0:
        even = even + 1
    else:
        odd = odd + 1
print("Number of even numbers : "+str(even))
print("Number of odd numbers : "+str(odd))
```

```python
Input = input("Sample numbers :  ")
List = Input.split(",")
even = [num for num in List if int(num)%2 == 0]
print("Number of even numbers : "+str(len(even)))
print("Number of odd numbers : "+str((len(List))-(len(even))))
```
